Imports

```python
import tensorflow
from tensorflow import keras
import matplotlib.pyplot as plt

from scipy import ndimage
```

Dataset load

```python
dataset = keras.datasets.fashion_mnist
((image_train, identification_train), (image_test, identification_test)) = dataset.load_data()
```

Data explore

```python
len(image_train)
image_train.shape
image_test.shape
len(identification_test)

identification_train.min()
identification_train.max()
```

Data show

```python
identification_quant = 10
description = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

for image in range(10):
    plt.subplot(2,5, image+1)
    plt.imshow(image_train[image], cmap='gray')
    plt.axis('off')
    plt.title(description[identification_train[image]])
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)), #flat the images
    keras.layers.Dense(256, activation=tensorflow.nn.relu), #negative input will habe zero as output
    keras.layers.Dense(10, activation=tensorflow.nn.softmax) #compare the image to every other classification
    ])

model.compile()

model.fit(image_train, identification_train)
```
